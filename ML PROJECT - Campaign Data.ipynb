```python
import numpy as np
import pandas as pd
df=pd.read_excel("/content/campaign_data_knovista_project (3).xlsx")
df
```

Check shape,size,null values and data types

```python
print(df.shape)
print(df.size)
print(df.isna().sum())
print(df.dtypes)
```

Input

```python
x=df.drop(['Campaign','Sales'],axis=1).values
x
```

Output

```python
y=df['Sales'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rf=RandomForestRegressor()
p={'n_estimators':[50,100,150],'max_depth':[3,5,7,9,11]}
gd=GridSearchCV(rf,p,cv=10,scoring='r2')
gd.fit(x_train,y_train)
print(gd.best_params_)
```

```python
rf1=RandomForestRegressor(n_estimators=150,max_depth=11)
rf1.fit(x_train,y_train)
y_pred=rf1.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print(mean_absolute_percentage_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
